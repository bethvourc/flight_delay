```python
import pandas as pd
import numpy as np 
import xgboost as xgb
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Import dataset

file_path = '/Users/bethvour/projects/flight_delay/Flight_delay.csv'

# Read the dataset

df = pd.read_csv(file_path)
df.head()
```

```python
# Select column from dataset
# CarrierDelay column represents delay by airline carrier itself

df = df[['DayOfWeek', 'Date', 'DepTime', 'Airline', 'Origin', 'Dest', 'CarrierDelay']]
df
```

```python
# Check for missing data from the new streamlined dataset

df.isnull().sum()
```

```python
# Convert the datetime date to pandas datetime
# Create df is a copy, since df is a slice from another Dataframe 

df = df.copy()
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
# Extract the month and the day 

df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
```

```python
# Drop the original 'Date' column as XGBoost doesn't handle datetime types 

df = df.drop(columns=['Date'])
```

```python
# The XGBoost library does not allow categorical data into the model, so these features 
# would have to be encoded into dummy variables AKA "one-hot encoding"

categories = df.select_dtypes(include=['object']).columns
categories 
```

```python
# One-Hot Encoding (Dummy variable) for Categoric Data

df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
# Recode carrier delay column into a new binary encoded variable as our target, prepare target 
# variable 

df_encoded['is_delayed_60+'] = np.where(df_encoded['CarrierDelay'] > 60, 1, 0)
```

```python
# Define features as an object called x and target variable as object called y 
X = df_encoded.drop(columns=['is_delayed_60+', 'CarrierDelay'])
y = df_encoded['is_delayed_60+']
```

```python
# Split the data into training and testing sets 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Exploratory Data Analysis - perform after train/test split to avoid data leakage (information from test set would influence the model)

# Put train X and y together - use the pandas Concat function to reconnect the X features with the y variable

train_set = pd.concat([X_train, y_train], axis = 1)
```

```python
# Undo one-hot encoding 

categories
```

```python
# Reversing one-hot encoding for multiple sets of categorical variables

for category in categories:
    one_hot_columns = [col for col in train_set.columns if col.startswith(f'{category}_')]
    train_set[category] = train_set[one_hot_columns].idxmax(axis=1)
    train_set = train_set.drop(columns=one_hot_columns)
    train_set[category] = train_set[category].str.replace(f'{category}_', '')
```

```python
train_set
```

```python
# Check distribution of target variable (number of flights that were delayed for over 60 minutes)

train_set['is_delayed_60+'].value_counts()
```

```python
# Calculate the percentage of training data that were delayed 

train_set['is_delayed_60+'].mean()
```

Dealing with class imbalance
1. Over-sampling smaller class
2. Under-sampling larger class
3. Specialized models

```python
# Delays by Airline

train_set.groupby('Airline')['is_delayed_60+'].mean().sort_values(ascending=False).round(3)*100
```

```python
# Delays by Day of the week

DayOfWeek_pct_delayed = train_set.groupby('DayOfWeek')['is_delayed_60+'].mean().round(3)*100
DayOfWeek_pct_delayed
```

```python
# Delays by Origin Airport (which origin airport has the highest chance of a delay)

pct_delay_by_origin = train_set.groupby('Origin')['is_delayed_60+'].mean().sort_values(ascending=False).round(3)*100
pct_delay_by_origin.head(20)
```

```python
# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(pct_delay_by_origin.values, bins=25, color='green', edgecolor='black')

# Add labels and title
plt.title("Distribution of 60+ Minute Delays By Origins", fontsize=14)
plt.xlabel("Percentage of 60+ Minute Delay (%)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Show the plot
plt.show()
```

```python
# Model Builiding

# Initalize and fit the XGBoost classifier (baseline model)

xgb_model = xgb.XGBClassifier(random_state=0, eval_metric='logloss')
```

```python
xgb_model
```

```python
# Train the model 

xgb_model.fit(X_train, y_train)
```

```python
# Predict on the test set

y_pred = xgb_model.predict(X_test)
```

```python
# Evaluate the model

print("XGBoost Classifier (Baseline):")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

```python
# Confusion matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

Note:- The bulk of delayed flights were predicted on the left cell of the matrix have been predicted as false negatives, but only 86 were accurately predicted, therefore, our high accuracy score of 92% is misleading not due to strong model performance but because the number of delayed flights are so low that the model does well by predicting that most flights are on time. 

```python
# Calculate the false positives and true positives to getting a better accuracy report of the prediction

# Predict probabilities for the test set (to calculate the AUC)
y_pred_proba = xgb_model.predict_proba(X_test)[:, 1] # We need probabilities for the positive class

# Calculate the Area Under the curve (AUC)
auc_score = roc_auc_score(y_test, y_pred_proba)
print(f"AUC Score: {auc_score:.4f}")
```

```python
# Plot the ROC curve

fpr, tpr, threshold = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC = {auc_score:.4f}')
plt.plot([0, 1], [0, 1], 'k--') # Diagonal line for random guessing 
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Postive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
# Do Cross-Validation Grid Search 

param_grid = {
    'learning_rate': [0.01, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 250],
    'subsample': [0.6, 1.0]
}

# Initialize the XGBoost model 

xgb_model = xgb.XGBClassifier(random_state=0,
                             eval_metric='logloss')

# Set up GridSearchCV

grid_search=GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    cv=3,
    scoring='roc_auc',
    verbose=1,
    n_jobs=-1)

# Fit the grid search model 

grid_search.fit(X_train, y_train)
```

```python
# Evaluate best parameters and predict on training set 

# Best parameters from GridSearch
print("Best parameters found: ", grid_search.best_params_)

# Predict with the best model
y_pred_best = grid_search.best_estimator_.predict(X_test)

# Evaluate the tuned XGBoost model 
print("XGBoost Classifier (Tuned):")
print(f"Accuracy: {accuracy_score(y_test, y_pred_best):.4f}")
```

```python
# Confusion matrix

cm = confusion_matrix(y_test, y_pred_best)
cm
```

```python
# Predict probabilities for the test set (to calculate AUC)
y_pred_proba = grid_search.best_estimator_.predict_proba(X_test)[:, 1] # Need the probabilities for the positive class

# Calculate the AUC score 
auc_score = roc_auc_score(y_test, y_pred_proba)
print(f"Auc Score: {auc_score:.4f}")
```

```python
# Plot the ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label=f'AUC = {auc_score:.4f}')
plt.plot([0,1], [0,1], 'k--') # Diagonal line for random guessing
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
